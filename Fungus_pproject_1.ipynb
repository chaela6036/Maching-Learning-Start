### MUSHROOM 데이터 가공해보기

[1]독립변수. 종속 변수 결정하기

```python
# 만약 독성과 식용의 비율이 크게 차이가 날 경우, 분류에 편향이 생길 가능성이 있음

#fungus_df에서 'class' 컬럼과 나머지 모든 컬럼 간의 교차표(class-tabulation) 출력

import pandas as pd
import numpy as np

for col in fungus_df.columns:
    if col != 'class':  # class 변수 제외
        print(f"=== class vs {col} ===")
        print(pd.crosstab(fungus_df['class'],fungus_df[col]))
        print('\n')
```
