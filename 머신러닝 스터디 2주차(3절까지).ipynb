### 첫번째 머신러닝 만들어 보기 - 붓꽃 품종 예측하기

```python
#붓꽃 데이터를 불러오기
from sklearn.datasets import load_iris
```

```python
#모듈을 기반으로 학습 (클래스 적용 후 의사결정 나무를 사용할 수 있도록)
from sklearn.tree import DecisionTreeClassifier
```

```python
#데이터 세트를 학습 데이터와 테스트 데이터로 분리 하는데 train_test_split 함수 (4절과 관련)
from sklearn.model_selection import train_test_split
```

```python
import pandas as pd

#iris data 불러오기 (Iris 데이터 세트에서 feature 만으로 된 데이터를 numpy.ndarray로 갖고 있음)
iris = load_iris()
iris_data= iris.data

print(type(iris_data))
print(iris_data)
```

```python
#iris target 불러오기
iris_label=iris.target
print('iris target 값:\n' ,iris_label)
print('iris target 명:\n' ,iris.target_names)
print(type(iris.target))
```

```python
#붓꽃 데이터 세트를 데이터 프레임으로 변환
iris_df = pd.DataFrame(data=iris_data, columns= iris.feature_names)


#label이라는 새 컬럼을 만든다.
iris_df['label']=iris.target
iris_df.head()

```

```python
#학습용 데이터와 테스트 용 데이터 분리 (n%가 테스트 데이터)
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2, random_state=11)
```

```python
# 객체 생성이란? DecisionTreeClassifier라는 설계도(클래스)로 dt_clf라는 도구(객체)를 만드는 것과 같음

dt_clf=DecisionTreeClassifier(random_state=11)
```

```python
#fit : 모델을 학습시키는 함수
#학습수행

dt_clf.fit(X_train, y_train)
```

```python
#학습이 완료된 DecisionTreeClassifier 객체에서 테스트 데이터 세트로 예측 수행
#예측은 학습 데이터가 아닌 것으로 실행 

pred = dt_clf.predict(X_test)
```

```python
#예측 결과를 기반으로 예측 성능 평가(정확도 측정)
from sklearn.metrics import accuracy_score
print('예측정확도 : {0:.4f}'.format(accuracy_score(y_test,pred)))
```
